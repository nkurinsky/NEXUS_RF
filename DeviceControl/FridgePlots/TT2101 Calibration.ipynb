```python
import sys, os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
cal_file = "ThermCal_TT2101.dat"
cal_path = "./"
cal_data = pd.read_csv(os.path.join(cal_path,cal_file),sep='\t',index_col=False,header=3,
                       names=["T_mK","R_TT2101","R_TT2102","R_TT2103"])
cal_data
```

```python
ax = plt.figure(figsize=(8,6)).gca()
ax.set_xlabel(r"Resistance [k$\Omega$]")
ax.set_ylabel("Temperature [mK]")

ax.plot(cal_data["R_TT2101"]/1e3,cal_data["T_mK"],label="TT2101")
ax.plot(cal_data["R_TT2102"]/1e3,cal_data["T_mK"],label="TT2102")
ax.plot(cal_data["R_TT2103"]/1e3,cal_data["T_mK"],label="TT2103")

ax.set_yscale('log')
ax.legend(loc='best')
```

```python
TT2101_cal     = interp1d(cal_data["R_TT2101"],cal_data["T_mK"])
TT2101_cal_inv = interp1d(cal_data["T_mK"],cal_data["R_TT2101"])
```
